```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
duration_df = pd.read_excel(r"C:\Users\rlampre\Desktop\ATO animal trial Durations.xlsx")
```

```python
print(duration_df.head())
```

```python
average_duration = duration_df.groupby(['AnimalID','Condition'], as_index=False)['Duration'].mean()

print(average_duration)
```

```python
pivot_df = average_duration.pivot(index='AnimalID', columns='Condition', values='Duration')
pivot_df = pivot_df[['Saline','CNO']]

print(pivot_df)
```

```python

plt.plot(pivot_df.columns, pivot_df.loc['ATO1'], marker='o', color='black', label='ATO1')
plt.plot(pivot_df.columns, pivot_df.loc['ATO6'], marker='o', color='gray', label='ATO6')

plt.xlabel('Condition')
plt.ylabel('Average Duration')
plt.title('Average Duration of ATO Trials by Animal and Condition')
plt.legend()
plt.show()
```
